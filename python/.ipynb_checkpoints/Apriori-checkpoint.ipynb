```python
#Importing the Libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the Dataset
fileName='/home/xenon/extensomllearn/Machine Learning A-Z/Part 5 - Association Rule Learning/Section 28 - Apriori/Apriori-Python/Market_Basket_Optimisation.csv'
dataset=pd.read_csv(fileName,header=None)
dataset.head()
```

```python
#Converting the dataset as Lists of List whcih is the format in which the function aspects the dataset.
transactions=[]
```

```python
for i in range(0,len(dataset)):
    transactions.append(str(dataset.values[i,j] for j in range(0,20)))
```

```python
#Fitting the Dataset in the apriori algorithm
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
#visualising the Rules
results=list(rules)
```
